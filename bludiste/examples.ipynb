```python
slalom = np.ones((20, 20), dtype = int)
slalom[0, :] = 0
slalom[0:6, 19] = 0
slalom[5, :] = 0
slalom[5:16, 0] = 0
slalom[15, :] = 0
slalom[15:20, 19] = 0
slalom = slalom.reshape((1, 400))
velikost = 20
```

```python
bludiste.generovani_bludiste_sablona(slalom, velikost)
bludiste.transformace_na_incidencni_matici()
bludiste.najdi_nejkratsi_cestu()
bludiste.mark_path_in_maze()
bludiste.vykresleni_grafu()
```
